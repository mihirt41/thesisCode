```python
# Import packages
from bs4 import BeautifulSoup
import requests
import json
import urllib2
import pandas as pd
```

```python
# Converting html response to dictionary formatted in json
speechLink = "https://www.federalreserve.gov/json/ne-speeches.json"
content = urllib2.urlopen(speechLink).read()
text = content[4:(len(content)-2)]
jsonData = json.loads(text)
```

```python
# Converting to dataframe
df = pd.DataFrame.from_dict(jsonData)
```

```python
# Renaming Columns
df.columns = ['RevisionNote', 'DateTime', 'Link','Location', 
                           'o', 'SpeakerAndPosition', 'SpeechTitle', 'UpdateDate',
                           'videoLink', 'IsVideo']
```

```python
# Scraping 2005 Speech data
speechLink2005 = "https://www.federalreserve.gov/newsevents/speech/2005speech.htm"
response = requests.get(speechLink2005, "html.parser")
pageContent = BeautifulSoup(response.content, "html.parser")
```

```python
# Creating lists of relevant data on each speech
titleDivs = pageContent.find_all("div", {"class":"title"})
links = []
titles = []
titleDivs
for titleDiv in titleDivs:
    # adding all links
    a = titleDiv.find_all("a", href = True)
    a = a[0]
    links.append(a['href'])
    
    # adding all titles
    titles.append(titleDiv.text)
    
speakerDivs = pageContent.find_all("div", {"class:"})
```
