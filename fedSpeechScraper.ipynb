```python
# Functionalizing previous code

def scrapeFedSpeech(link):
    response = requests.get(link, "html.parser")
    pageContent = BeautifulSoup(response.content, "html.parser")
    titleDivs = pageContent.find_all("div", {"class":"title"})
    links = []
    titles = []
    titleDivs
    for titleDiv in titleDivs:
        # adding all links
        a = titleDiv.find_all("a", href = True)
        a = a[0]
        links.append(a['href'])
        # adding all titles
        titles.append(titleDiv.text)

    speakerDivs = pageContent.find_all("div", {"class":"speaker"})
    speakers = []
    for speakerDiv in speakerDivs:
        speaker = speakerDiv.text
        speakers.append(speaker)
       

    locationDivs = pageContent.find_all("div", {"class":"location"})
    locations = []
    for locationDiv in locationDivs:
        location = locationDiv.text
        locations.append(location)
    

    dates = []
    ulDiv = pageContent.find_all("ul", {"id":"speechIndex"})
    liDivs = ulDiv[0].find_all("li")
    for liDiv in liDivs:
        date = liDiv.contents[0]
        dates.append(date)
    

    df= pd.DataFrame({
    'speaker':speakers,
    'date':dates,
    'location':locations,
    'title':titles,
    'link':links
    })
    
    return df
    
```

```python
dfs = []
for i in range(1996,2006):
    year = str(i)
    speechLink = "https://www.federalreserve.gov/newsevents/speech/" + year + "speech.htm"
    df = scrapeFedSpeech(speechLink)
    dfs.append(df)
```

```python
# Getting Statement and minutes data

def releaseScraper(yearLink):
    response = requests.get(yearLink,"html.parser")
    pageContent =  BeautifulSoup(response.content, "html.parser")

    panelList = pageContent.find_all("div", {"class":"col-xs-12 col-md-6"})

    policyLinks = []
    for panel in panelList:
        a = panel.find_all("a", {'href': re.compile(r'newsevents/pressreleases/monetary')})
        if a:
            policyLinks.append(a)

    minutesLinks = []
    for panel in panelList:
        a = panel.find_all("a", {'href': re.compile(r'monetarypolicy/fomcminutes')})
        if a:
            minutesLinks.append(a)

    newMinutesLinks = []
    for link in minutesLinks:
        link = link[0]
        newMinutesLinks.append("https://www.federalreserve.gov" + link['href'])

    newPolicyLinks = []
    for policyLink in policyLinks:
        policyLink = policyLink[0]
        newPolicyLinks.append("https://www.federalreserve.gov" + policyLink['href'])

    statementDates = pageContent.findAll(text = re.compile(r'Meeting - '))
    minutesDates = pageContent.findAll(text = re.compile(r'Minutes '))

    dfStatement = pd.DataFrame({
        'statementLink':newPolicyLinks,
        'statementDate' :statementDates
    })
    
    dfMinutes =  pd.DataFrame({
        'minutesLink':newMinutesLinks,
        'minutesDate' :minutesDates
    })
    
    return dfStatement, dfMinutes
```

```python
searchLinks = []
for i in range(2010, 2013):
  searchLinks.append("https://www.federalreserve.gov/monetarypolicy/fomchistorical" + str(i)+".htm")  

statementDfs = []
minutesDfs = []
for searchLink in searchLinks:
    print(searchLink)
    dfs = releaseScraper(searchLink)
    statementDfs.append(dfs[0])
    minutesDfs.append(dfs[1])
    
```
